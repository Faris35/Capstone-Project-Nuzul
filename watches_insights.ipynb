```python
# Function to remove outliers for each brand
def remove_outliers_by_brand(df, column='price_usd', group_col='brand', lower_percentile=0.01, upper_percentile=0.99):
    def filter_outliers(group):
        lower_bound = group[column].quantile(lower_percentile)
        upper_bound = group[column].quantile(upper_percentile)
        return group[(group[column] >= lower_bound) & (group[column] <= upper_bound)]
    
    return df.groupby(group_col, group_keys=False).apply(filter_outliers)

# Apply the function to the dataset
df = remove_outliers_by_brand(df)
```

# Are certain materials (e.g., gold, leather) more common in specific price ranges?

```python
# Helper function to process distributions and ensure consistent colors
def prepare_and_filter_consistent_colors(data, top_n=5):
    # For each row, keep top_n materials and aggregate the rest as 'Other'
    processed_data = data.copy()
    for index, row in processed_data.iterrows():
        top_materials = row.nlargest(top_n).index
        processed_data.loc[index, 'Other'] = row.drop(top_materials).sum()
        processed_data.loc[index, ~processed_data.columns.isin(top_materials)] = 0
    # Drop columns with all zeros
    processed_data = processed_data.loc[:, (processed_data != 0).any(axis=0)]
    return processed_data

# Prepare and filter the data for case and bracelet materials
filtered_case_material = prepare_and_filter_consistent_colors(case_material_price_distribution, top_n=5)
filtered_brace_material = prepare_and_filter_consistent_colors(brace_material_price_distribution, top_n=5)

# Define consistent colors for materials
all_materials = list(set(filtered_case_material.columns) | set(filtered_brace_material.columns))
material_colors = {material: color for material, color in zip(all_materials, plt.cm.tab20.colors)}

# Helper function to create pie charts with consistent colors
def pie_chart_with_consistent_colors(distribution, title):
    fig, axes = plt.subplots(1, len(distribution.index), figsize=(20, 6), subplot_kw=dict(aspect='equal'))
    for ax, (index, row) in zip(axes, distribution.iterrows()):
        sizes = row.values
        labels = row.index
        # Filter out zero values
        filtered_sizes = [size for size in sizes if size > 0]
        filtered_labels = [label for size, label in zip(sizes, labels) if size > 0]
        filtered_colors = [material_colors[label] for label in filtered_labels]
        ax.pie(
            filtered_sizes,
            labels=filtered_labels,
            colors=filtered_colors,
            autopct='%1.1f%%',
            textprops={'fontsize': 8},
        )
        ax.set_title(index, fontsize=10)
    fig.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.show()

# Visualize case material distribution with consistent colors
pie_chart_with_consistent_colors(filtered_case_material, 'Case Material Distribution Across Price Ranges')

# Visualize bracelet material distribution with consistent colors
pie_chart_with_consistent_colors(filtered_brace_material, 'Bracelet Material Distribution Across Price Ranges')
```

# How does the condition of a watch affect its price?

```python
# Create a creative boxplot for price distribution by condition
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='condition', y='price_usd', palette='coolwarm')
plt.title('Price Distribution by Condition', fontsize=16)
plt.xlabel('Condition', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.ylim(0, df['price_usd'].quantile(0.95))  # Focus on 95% of data to reduce the impact of extreme outliers
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

1. **Condition and Price Range**:
   - Watches in **"Unworn"** condition have the highest median price, suggesting they are considered the most valuable.
   - **"New"** condition watches have a lower median price compared to "Unworn" but are still highly valued, indicating that even minor usage impacts pricing.
   - Watches in **"Very Good"** condition also maintain relatively high prices, with a median slightly below "New."

2. **Spread of Prices**:
   - The price distribution for **"Very Good"**, **"New"**, and **"Unworn"** watches shows a wider range, indicating that these conditions cater to both mid-range and high-end luxury markets.
   - In contrast, conditions like **"Fair"** and **"Poor"** have narrower price distributions, reflecting limited market interest and lower overall value.

3. **Outliers**:
   - Significant outliers are observed in the **"New"** and **"Unworn"** categories, with prices exceeding $50,000. These likely represent ultra-luxury or rare models that retain high value due to exclusivity or craftsmanship.
   - Few outliers are present in **"Fair"** and **"Poor"** conditions, underscoring limited demand for watches in these categories.

4. **"Poor" Condition Watches**:
   - Watches in **"Poor"** condition have the lowest prices, with almost no variation. This suggests that such watches are considered functional at best, with minimal collector or resale value.

5. **Market Implications**:
   - **Condition strongly influences pricing**, with "Unworn" and "New" watches commanding higher prices due to perceived quality and desirability.
   - Watches in "Fair" and "Poor" condition are less competitive in the luxury market, catering mainly to buyers seeking affordability or restoration opportunities.

### Key Takeaway:
The graph clearly illustrates the **strong correlation between condition and price** in the luxury watch market. Buyers prioritize watches in "Unworn" or "New" condition, while "Very Good" watches remain a viable option at slightly lower price points. Conversely, "Fair" and "Poor" condition watches struggle to maintain value due to reduced desirability.
